```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_loss(df, title):
    fig, ax = plt.subplots(figsize=(7, 4.5))
    ax.plot(df["epoch"], df["train_loss"], color="blue", marker="o", markersize=3, label="train")
    ax.plot(df["epoch"], df["val_loss"], color="orange", marker="o", markersize=3, label="val")
    ax.set_title(title)
    ax.set_xlabel("epoch")
    ax.set_ylabel("loss")
    ax.legend()
    ax.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
```
