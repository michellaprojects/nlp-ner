```python
import pandas as pd
from pandas import json_normalize
```

```python
df = pd.read_json('../datasets/EntityRecognitioninResumes.json', lines=True)
pd.set_option('display.max_colwidth', None)
df.head(5)
```

```python
print(df.columns)
df['extras'].unique
```

```python
df = df.drop(columns='extras')
print(df.columns)
```

```python
first_resume_annotations = df['annotation'].iloc[0]
eda_friendly_df = json_normalize(first_resume_annotations)
print(eda_friendly_df.head())
```

```python
all_data = []

for index, row in df.iterrows():
    
    for ann in row['annotation']:
        label = ann['label'][0] if ann['label'] else "None"
      
        text = ann['points'][0]['text']
        
        all_data.append({
            'resume_id': index, 
            'label': label, 
            'text': text.strip()
        })


clean_df = pd.DataFrame(all_data)


print(clean_df['label'].value_counts())
```
